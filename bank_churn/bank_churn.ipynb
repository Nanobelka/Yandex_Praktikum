```python
# загрузка данных

def df_read_csv(path, sep=','):
    
    try:
        return pd.read_csv(path, sep)
    except Exception:
        print('Ошибка при загрузке файла')


#--------------------------------------------------------------------
# определение имени таблицы

def df_name(df):
    
    return [name for name in globals() if globals()[name] is df][0]


#--------------------------------------------------------------------
# первичная информация о таблице

def basic_info(df, sample='sample', qnty=10):

    print(f'\n{f.BOLD}таблица {df_name(df)}{f.END} {theme_char*20}\n\n')

    # метод info()
    print(df.info())

    # несколько записей
    if sample == 'sample':
        display(df.sample(qnty))
    elif sample == 'head':
        display(df.head(qnty))
    elif sample == 'tail':
        display(df.tail(qnty))
    
    # метод describe()
    print(df.describe(), '\n')
    print(df.describe(exclude=['int', 'int64', 'uint', 'uint64', 'float', 'float64']), '\n')


#--------------------------------------------------------------------
# перевод значений полей типа object в нижний регистр

def object_fields_lower(df):  # dataframe
    
    for i in range(len(df.dtypes)):
        if df.dtypes[i] == 'object':
            df[df.columns[i]] = df[df.columns[i]].str.lower()


#--------------------------------------------------------------------
# проверка на дубликаты

def duplicated_test(df):
    print('\n',
          f'количество полных дубликатов в таблице {f.BOLD + df_name(df) + f.END}: {df.duplicated().sum()}',
          '\n')


#--------------------------------------------------------------------
# числовые значения для вертикальных bar-графиков
# доработать до универсальной (+ горизонтальные)

def bar_annotate(ax, decimals=0):
    
    max_height = 0
    for p in ax.patches:
        if p.get_height() > max_height:
            max_height = p.get_height()
    
    for p in ax.patches:
        _x = p.get_x() + p.get_width() / 2
        _y = p.get_y() + p.get_height() + max_height*0.01
        value = f'{p.get_height():.{decimals}f}'
        ax.text(_x, _y, value, ha="center")
```

### Cредние значения признаков, сгруппированных по полю `churn`

Сравним средние значения признаков для оставшихся и ушедших клиентов.

```python
# ratio показывает коэффициент различия между средними значениями признаков для churn=0 и churn=1
# ratio всегда >= 1

def ratio(df):
    if df.active == 0 or df.churn == 0:
        return 0
    if df.active > df.churn:
        return df.active / df.churn
    else:
        return df.churn / df.active


features = (
            data
            .drop('user_id', axis=1)
            .groupby('churn')
            .mean()
            .T
            .sort_index()
            .rename(columns={0:'active', 1:'churn'})
           )

features['ratio'] = features.apply(ratio, axis=1)


space()
print('Cредние значения признаков, сгруппированных по полю `churn`')
print('(чем выше `ratio`, тем сильнее различаются `active` и `churn`)')

display(
        features
        .sort_values('ratio', ascending=False)
        .style
              .background_gradient('Reds', subset='ratio', high=1)
              .format('{:.3f}')
       )

# сохранение признаков в виде списка, отсортированного по предполагаемой силе связи с полем churn
# далее этот список будет использован при построении таблиц и графиков
features_sorted = (
                   features
                   .sort_values('ratio', ascending=False)
                   .index
                   .to_list()
                  )
```

> Для оценки силы влияния признаков на отток клиенты разбиты на 2 группы: `active` и `churn`, для которых рассчитаны средние значения признаков.
>

## Кластеризация клиентов

Поскольку установленное влияние признаков на отток не является ярко выраженным, проведем дополнительное исследование. Возможно, обнаружится значимое влияние на отток клиентов и среди других признаков.

Для этого проведем кластеризацию по алгоритму K-Means и оценим влияние признаков при делении на кластеры.

**Важно!** Кластеризация осуществляется не с целью использования полученных кластеров для работы банка, а с целью выявления влияния признаков на отток клиентов.

### Функция для кластеризации

Задача функции: разделить клиентов на число кластеров от 2-х до заданного максимума, для каждого варианта показать долю оттока и средние значения всех признаков.

Входные данные функции:  
    `x` — датафрейм с признаками (без целевой переменной);  
    `max_clasters` — максимальное количество кластеров при построении моделей K-Means.

```python
def clasterization(x, max_clasters):

    # создание и обучение стандартизатора, стандартизация данных
    x_st = StandardScaler().fit_transform(x)

    # матрица расстояний
    linked = linkage(x_st, method='ward')

    # дендрограмма
    plt.figure(figsize=(14,4))  
    dendrogram(linked, orientation='top', no_labels=True)
    plt.title('Дендрограмма кластеризации клиентов', size=20, pad=30)
    space()
    plt.show()
    
    # упорядоченный список полей (целевая переменная churn + признаки)
    # чтобы наиболее важные признаки при выводе стояли ближе к левому краю таблицы
    fields_sorted = ['churn'] + features_sorted
    
    # создание моделей K-Mens
    
    for i in range(2, max_clasters+1):

        # создание и обучение модели K-Means
        # кластеризация данных
        # добавление инфо о кластеризации в общий dataframe
        data['KM_cluster'] = KMeans(n_clusters=i, n_init=10, random_state=0).fit_predict(x_st)


        # таблица средних значений, сгруппированная по кластерам
        display(
                data
                .groupby('KM_cluster')
                [fields_sorted]       # перестановка признаков в порядке убывания их предполагаемой связи с churn
                .mean()
                .sort_values('churn', ascending=False)
                .replace({'city_id':{324:'Ростов'}})
                .style
                      .background_gradient('Reds', subset='churn', high=1)
#                       .format('{:.2f}')
                      .format('{:.1%}', subset='churn')
               )


        # оценка качества кластеризации метрикой silhouette
        silhouette = silhouette_score(x_st, data.KM_cluster)
        print(
              'Метрика Silhouette для модели K-Means из {} кластеров: {:.4f}'
              .format(i, silhouette)
             )
```

Поле `balance` содержит пропуски — 1/3 часть всех значений. Перед созданием моделей необходимо избавиться от пропусков. Это можно сделать одним из способов:
- заполнить пустые значения;
- удалить строки, содержащие пустые значения;
- полностью удалить поле `balance`.

Первый способ не подходит, т.к. нет уверенности, что NaN эквивалентно нулю, а заполнение расчетным значением неизбежно исказит результат.

Второй способ не годится, потому что будет утеряно большое количество записей.

Третий способ наиболее приемлем, особенно с учетом, что по результатам проведенных исследований не предполагается сколько либо значимой зависимости `churn` от `balance`.

### Кластеризация по широкому списку признаков

Для начала будет проведена кластеризация с использованием всех имеющихся признаков.

Максимальное количество кластеров, равное 6,  подобрано экспериментально.

## Проверка гипотез

По результатам проведенных исследований сформулирован ряд гипотез для проверки предварительных выводов:

- доли ушедших клиентов для разных значений loyalty различаются;
- доли ушедших клиентов среди мужчин и женщин различаются;
- доли ушедших клиентов в разных городах различаются;
- распределение возраста ушедших и оставшихся клиентов различается;

Кроме того проверим гипотезу, сформулированную  в задании:  
- распределение дохода ушедших и оставшихся клиентов различается.

Ввиду проведения множественных тестов на одном и том же наборе данных используем поправку Бонферрони.

### Определение уровня статистической значимости

```python
alpha_single = 0.05        # "традиционный" уровень статистической значимости для одиночного теста

N = 7                      # количество тестов

alpha = alpha_single / N   # уровень статистической значимости с поправкой Бонферрони
```

### Функция для Z-теста

Функция выводит результаты Z-теста.

Входные данные:  
A_total – общее количество клиентов в серии A;  
A_churn – количество ушедших клиентов в серии A;  
B_total – общее количество клиентов в серии B;  
B_churn – количество ушедших клиентов в серии B.

```python
def Z_test(A_total, A_churn, B_total, B_churn):
    
    space()
    
    # расчеты

    P0 = A_churn / A_total
    P1 = B_churn / B_total
    P = (A_churn + B_churn) / (A_total + B_total)

    z_value = (P0 - P1) / (P * (1 - P) * (1/A_churn + 1/B_churn)) ** (1/2)

    distr = stats.norm(0, 1)  # стандартное нормальное распределение (среднее 0, стандартное отклонение 1)

    p_value = (1 - distr.cdf(abs(z_value))) * 2


    # вывод результата

    print('p-value: {0:.12f}'.format(p_value))

    if (p_value < alpha):
        print('при исходном alpha={:0.2f} и поправке Бонферрони ={}'.format(alpha_single, N))
        print('разница статистически значима, нулевая гипотеза отвергнута')
    else:
        print('при исходном alpha={:0.2f} и поправке Бонферрони ={}'.format(alpha_single, N))
        print('вывод о различии сделать нельзя, не получилось отвергнуть нулевую гипотезу')
```

### Гипотеза: отток зависит от значения loyalty

Нулевая гипотеза: доли ушедших клиентов среди loyalty=0 и loyalty=1 равны.  
Альтернативная гипотеза: доли ушедших клиентов среди loyalty=0 и loyalty=1 различаются.

Проверка с помощью Z-критерия.

```python
# подготовка данных

df = (
      data.groupby('loyalty')
          .churn
          .value_counts()
          .to_frame()
          .rename(columns={'churn':'cnt'})
          .reset_index()
     )
display(df)


# Z-тест

Z_test(
       df.cnt[0] + df.cnt[1],   # общее количество клиентов   с loyalty=0
       df.cnt[1],               # количество ушедших клиентов с loyalty=0
       df.cnt[2] + df.cnt[3],   # общее количество клиентов   с loyalty=1
       df.cnt[3]                # количество ушедших клиентов с loyalty=1
      )
```

> Доли ушедших клиентов среди loyalty=0 и loyalty=1 различаются (гипотеза о равенстве отвергнута).

### Проверка выборки на соответствие нормальному распределению

Перед проверкой запланированных гипотез проверим выборки по полям `age` и `salary` на соответствие нормальному распределению. Для этого сформулируем следующие гипотезы: 

1.  
Нулевая гипотеза: распределение `age` нормально.  
Альтернативная гипотеза: распределение `age` не нормально.  

2.  
Нулевая гипотеза: распределение `salary` нормально.  
Альтернативная гипотеза: распределение `salary` не нормально.

Проверку гипотез выполним, применив критерий Шапиро-Уилка.

```python
def shapiro_wilk_test(series):
    
    space()
    
    alpha_SW = 0.05                      # для этого теста своя собственная альфа 
    
    p_value = stats.shapiro(series)[1]   # p-value — значение с индексом 1 в массиве результатов

    print('p-значение: ', p_value)

    if (p_value > alpha_SW):
        print(f'Не получилось отвергнуть нулевую гипотезу: распределение {f.BOLD}{series.name}{f.END} нормально')
    else:
        print(f'Отвергаем нулевую гипотезу: распределение {f.BOLD}{series.name}{f.END} не нормально')
```

```python
shapiro_wilk_test(data.age)         # для запланированной гипотезы
shapiro_wilk_test(data.salary)      # для запланированной гипотезы
shapiro_wilk_test(data.balance)     # для проверки работы функции
shapiro_wilk_test(data.score)       # для проверки работы функции
```

Для проверки гипотез с участием выборок с **нормальным** распределением подойдет параметрический тест Стьюдента.  
Для проверки гипотез с участием выборок с **ненормальным** распределением подойдет непараметрический тест Манна-Уитни.

### Функция для теста Стьюдента либо теста Манна-Уитни

Функция определяет наличие статистической разницы между двумя последовательностями, используя заказанный тип теста.

Входные данные:  
test_type – типа теста: Стьюдента или Манна-Уитни;  
ser_A – первая последовательность;  
ser_B – вторая последовательность.

Тест Стьюдента в данной работе не используется. В функцию он введен как задел на создание универсальной функции для проверки гипотез.

```python
def hyp_test(test_type, ser_A, ser_B):

    if test_type == 'ttest_ind':
        p_value = stats.ttest_ind(ser_A, ser_B, equal_var=False).pvalue
        
    elif test_type == 'mannwhitneyu':
        p_value = stats.mannwhitneyu(ser_A, ser_B).pvalue


    # вывод результата

    space()

    print('p-value: {0:.12f}'.format(p_value))

    if (p_value < alpha):
        print('при исходном alpha={:0.2f} и поправке Бонферрони ={}'.format(alpha_single, N))
        print('разница статистически значима, нулевая гипотеза отвергнута')
    else:
        print('при исходном alpha={:0.2f} и поправке Бонферрони ={}'.format(alpha_single, N))
        print('вывод о различии сделать нельзя, не получилось отвергнуть нулевую гипотезу')
```

    - предпенсионная программа банка больше ориентирована на мужчин;
    - банк потерял крупного клиента, а с ним и его сотрудников, которые по совпадению оказались немолодыми;
    - банк имеет особую программу по удержанию молодых, и это их отток ниже нормального, а отток немолодых хоть и выше, но в пределах нормы.
1. Разобраться, почему мужчины в целом более лояльны, чем женщины. Возможно:
    - женщин оскорбляет сексизм в рекламе банка;
    - женщинам труднее ориентироваться в приложении или на сайте;
    - подарки к 23-му февраля были весомее, чем к 8-му марта.
1. Разобраться, что происходит в Ростове. Возможно:
    - конкуренты жестко демпингуют;
    - в городе ходят слухи, что банк закроется;
    - в городе закрылось крупное предприятие и "всех выгнали на мороз".
1. Разобраться, что означает поле loyalty. Возможно:
    - это клиенты, которые не замечают роста тарифов, в то время как другие замечают;
    - это клиенты, которым банк пообещал дать прибавку по вкладам и отобрать, если клиент уйдет;
    - это родственники сотрудников банка, которых уволят, если родственник уйдет.
