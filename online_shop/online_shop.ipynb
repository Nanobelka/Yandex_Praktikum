# Исследование данных интернет-магазина "Озозан"

**Заказчик:** интернет-магазин "Озозан".

**Данные:**  
- список гипотез, предоставленный отделом маркетинга;
- данные о заказах за период с 01.08.2019 по 31.08.2019;
- данные о посетителях за период с 01.08.2019 по 31.08.2019.

**Цель:** подготовить рекомендации по результатам анализа А/В-теста.

**Задачи:**

- приоритизировать гипотезы, подготовленные отделом маркетинга;
- проанализировать полученные результаты А/В-теста:
    - оценить качество предоставленных данных;
    - провести подготовку данных к анализу;
    - проанализировать по группам: конверсию, выручку, среднюю сумму заказа;
    - проанализировать распределение: количества заказов на одного покупателя, стоимость заказов;
    - попробовать найти различие между группами А и В с помощью статистических методов;
    - определить аномалии в данных, подготовить очищенные от аномалий данные;
    - при необходимости повторить анализ на основании очищенных данных;
- дать рекомендации заказчику по результатам исследования;
- определить потенциальные риски данного исследования, возможности их устранения;  
- дать рекомендации по дальнейшему развитию данного исследования.

### Подготовка среды

```python
import pandas as pd
import numpy as np
from scipy import stats
import random
import warnings
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
# ПОЛЕЗНЫЕ ФУНКЦИИ

#--------------------------------------------------------------------
# определение имени таблицы
def df_name(df):
    return [name for name in globals() if globals()[name] is df][0]

#--------------------------------------------------------------------
# вывод таблиц с выравниванием полей типа obect по левому краю
def display_left(df):
    
    object_type_fields = []
    for column in df.select_dtypes('object').columns:   # составление списка полей типа object
        object_type_fields.append(column)
    
    # выравнивание значений
    left_aligned_df = df.style.set_properties(
                                              subset=object_type_fields,
                                              **{'text-align': 'left'}
                                             )
    # выравнивание заголовков
    left_aligned_df = left_aligned_df.set_table_styles(
                                                       [dict(subset=object_type_fields,
                                                             selector='th',
                                                             props=[('text-align', 'left')]
                                                            )
                                                       ]
                                                      )
    return left_aligned_df
```

```python
# ДЛЯ ОФОРМЛЕНИЯ

#--------------------------------------------------------------------
warnings.filterwarnings('ignore')            # чтобы не портить внешний вид
pd.set_option('display.precision', 3)
pd.set_option('display.max_colwidth', 150)

#--------------------------------------------------------------------
# для свободного пространства (пустые строки)

def space(num=1):
    print('\n' * num)
    
#--------------------------------------------------------------------
# для украшения и разнообразия: ✺ ★ ☆ ✦ ✻ ⚛ ∻ ⋰

theme_char = '∻'

#--------------------------------------------------------------------
# для визуального дополнения при выводе

def line_end(length):
    return theme_char * length

#--------------------------------------------------------------------
# для оформления текста
class font:
    BOLD = "\033[1m"
    END = "\033[0m"
```

## Общая информация о входных данных

```python
def basic_info(df, sample='sample', qnty=10):

    print(font.BOLD + '\n'*2 + f'таблица {df_name(df)}' + font.END + '   ' + line_end(40) + '\n')
    print(df.info())
    if sample == 'sample':
        display(df.sample(qnty))
    elif sample == 'head':
        display(df.head(qnty))
    else:
        display(df.tail(qnty))
    print(df.describe(), '\n')
    print(df.describe(exclude=['int', 'uint', 'float']))
```

```python
basic_info(hypothesis, 'head', 10)
basic_info(orders, 'sample', 5)
basic_info(visits, 'sample', 5)
```

```python
def duplicated_test(df):
    print(f'количество полных дубликатов в таблице {df_name(df)}: {df.duplicated().sum()}')
```

```python
duplicated_test(hypothesis)
duplicated_test(orders)
duplicated_test(visits)
```

```python
# проверка по сокращенному количеству (основных) полей

orders_cuted = orders[['id_order']]
visits_cuted = visits[['date', 'group']]

duplicated_test(orders_cuted)
duplicated_test(visits_cuted)
```

### Другие проверки

## A/B-тест

### Кумулятивная выручка по группам

```python
# подготовка данных

# объединение агрегированных orders и visits
aggregation = pd.merge(
                       (
                        orders
                        .groupby(['group', 'date'])
                        .agg({'id_order':'nunique', 'id_user':'nunique', 'revenue':'sum'})
                        .rename(columns={'id_order':'orders', 'id_user':'buyers'})
                        .reset_index()
                       ),
                       visits,
                       on=['group', 'date'],
                       how='left'                          # для подстраховки
                      )

# создание кумулятивных данных (накопление по дате раздельно для групп)
aggregation_cumulative = (
                          aggregation
                          .set_index(['date','group'])
                          .sort_values(['date','group'])   # для подстраховки
                          .groupby(level=[1])              # группировка по полю ['group']
                          .cumsum()
                          .reset_index()
                         )

display(aggregation_cumulative.query('group == "A"').head())
display(aggregation_cumulative.query('group == "B"').head())
space()
```

```python
# график кумулятивной выручки

fig, ax = plt.subplots(figsize=(16,6), dpi=150)

A = aggregation_cumulative.query('group == "A"')
B = aggregation_cumulative.query('group == "B"')

ax.plot(A['date'], A['revenue'], label='группа A', linewidth=5, alpha=0.4, color='blue')
ax.plot(B['date'], B['revenue'], label='группа B', linewidth=5, alpha=0.4, color='red')
ax.set_title('Кумулятивная выручка по группам', fontsize=20, pad=25)
ax.set_xlabel('дата наблюдения', fontsize=12, labelpad=20)
ax.set_ylabel('выручка, руб.', fontsize=12, labelpad=20)
ax.legend(loc='upper left', fontsize=16)
ax.set_ylim(0,)
ax.grid(True, linestyle='--')
ax.tick_params(labelsize=12)
ax.set_facecolor('#f9f9f9')

space()
plt.show()
```

>На графике **группы В** есть существенный скачок в одну из дат. Необходимо проверить, чем он вызван: аномально большим количеством заказов или одиночным крупным заказом.  
За исключением данного скачка графики обеих групп растут приблизительно линейно без серьезных отклонений.  
Предположительно разница в выручке между группами стабилизировалась и без учета скачка ее не было бы.

### Средний чек

#### Кумулятивный средний чек по группам

```python
# расчет кумулятивного среднего чека

aggregation_cumulative['order_average'] = aggregation_cumulative['revenue'] / aggregation_cumulative['orders']


# график кумулятивного среднего чека

fig, ax = plt.subplots(figsize=(16,6), dpi=150)

A = aggregation_cumulative.query('group == "A"')
B = aggregation_cumulative.query('group == "B"')

ax.plot(A['date'], A['order_average'], label='группа A', linewidth=5, alpha=0.4, color='blue')
ax.plot(B['date'], B['order_average'], label='группа B', linewidth=5, alpha=0.4, color='red')
ax.set_title('Кумулятивный средний чек по группам', fontsize=20, pad=25)
ax.set_xlabel('дата наблюдения', fontsize=12, labelpad=20)
ax.set_ylabel('средний чек, руб.', fontsize=12, labelpad=20)
ax.legend(loc='upper left', fontsize=16)
ax.set_ylim(0, 12000)
ax.grid(True, linestyle='--')
ax.tick_params(labelsize=12)
ax.set_facecolor('#f9f9f9')

space()
plt.show()
```

>У **группы А** на стабилизацию среднего чека ушло около двух недель.  
Средний чек **группы В** стабилизировался было раньше, но все испортил одиночный крупный заказ. Он же повлиял на график кумулятивной выручки.  
После серьезного всплеска, вызванного крупным заказом, средний чек в **группе В** все еще стабилизируется, продолжая снижение. По данному графику непонятно, сравняются ли средние чеки групп, или какая-либо из груп будет стабильно выше.

#### Относительное изменение среднего чека группы B к группе A

```python
# расчет относительного изменения среднего чека группы B к группе A

AB = pd.merge(
              aggregation_cumulative.query('group == "A"')[['date', 'order_average']],
              aggregation_cumulative.query('group == "B"')[['date', 'order_average']],
              on='date',
              how='outer',
              suffixes=['_A', '_B']
             )

AB['order_B_relative_A'] = AB['order_average_B'] / AB['order_average_A'] - 1

# график относительного изменения среднего чека группы B к группе A

fig, ax = plt.subplots(figsize=(16,6), dpi=150)

ax.plot(AB['date'], AB['order_B_relative_A'], label='группа B', linewidth=5, alpha=0.4, color='red')
ax.axhline(y=0, linestyle='-', xmin=0.045, xmax=0.955, linewidth=5, alpha=0.4, color='blue', label='группа A')
ax.set_title('Относительное изменение среднего чека группы B к группе A', fontsize=20, pad=25)
ax.set_xlabel('дата наблюдения', fontsize=12, labelpad=20)
ax.set_ylabel('дельта относительного изменения', fontsize=12, labelpad=20)
ax.legend(loc='upper left', fontsize=16)
ax.set_ylim(-0.5, 0.7)
ax.grid(True, linestyle='--')
ax.tick_params(labelsize=12)
ax.set_facecolor('#f9f9f9')

space()
plt.show()
```

>Данный график подтверждает, что около двух недель ушло на стабилизацию среднего чека групп относительно друг друга. Далее произошел выброс, который не позволяет достоверно судить о происходящей тенденции.

### Статистическая значимость

#### Статистическая значимость различий в конверсии между группами по «сырым» данным

Нулевая гипотеза: конверсии в группах A и B равны.  
Альтернативная гипотеза: конверсии в группах A и B различаются.

```python
# подготовка данных для сравнения по критерию Манна-Уитни

# группа A

A_orders = (                                                               # количество заказов для каждого id_user
            orders.query('group == "A"')
            .groupby('id_user', as_index=False)
            .agg({'id_order': 'nunique'})
            .rename(columns={'id_order':'orders_qnty'})
           )

A_gapers = visits.query('group == "A"')['visitors'].sum() - len(A_orders)  # количество зевак

A = pd.concat(
              [A_orders['orders_qnty'],
               pd.Series(0, index=np.arange(A_gapers), name='orders_qnty')
              ],
              axis=0
             )

# группа B

B_orders = (                                                               # количество заказов для каждого id_user
            orders.query('group == "B"')
            .groupby('id_user', as_index=False)
            .agg({'id_order': 'nunique'})
            .rename(columns={'id_order':'orders_qnty'})
           )

B_gapers = visits.query('group == "B"')['visitors'].sum() - len(B_orders)  # количество зевак

B = pd.concat(
              [B_orders['orders_qnty'],
               pd.Series(0, index=np.arange(B_gapers), name='orders_qnty')
              ],
              axis=0
             )
```

```python
# сравнение по критерию Манна-Уитни
# гипотеза-0: конверсии в группах A и B равны

alpha = 0.05
results = stats.mannwhitneyu(A, B)

space()

print('p-value: {0:.4f}'.format(results.pvalue))
print('отношение конверсий (группа B к группе A): {0:.3f}'.format(B.mean() / A.mean()))

if (results.pvalue < alpha):
    print('отвергаем нулевую гипотезу: разница статистически значима при alpha={}'.format(alpha))
else:
    print('не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя при alpha={}'.format(alpha))

space()
```

Альтернативная проверка той же гипотезы с использованием Z-критерия.  
Нулевая гипотеза: конверсии в группах A и B равны.  
Альтернативная гипотеза: конверсии в группах A и B различаются.

```python
Z = pd.merge(
             orders.groupby('group').agg({'id_order':'count'}).rename(columns={'id_order':'orders'}),
             visits.groupby('group').agg({'visitors':'sum'}).rename(columns={'visitors':'visits'}),
             on='group'
            )
display(Z)
```

```python
# расчет с использованием Z-критерия

Z = pd.merge(
             orders.groupby('group').agg({'id_order':'count'}).rename(columns={'id_order':'orders'}),
             visits.groupby('group').agg({'visitors':'sum'}).rename(columns={'visitors':'visits'}),
             on='group'
            )

p_A = Z.loc['A','orders'] / Z.loc['A','visits']
p_B = Z.loc['B','orders'] / Z.loc['B','visits']
p_AB = (Z.loc['A','orders'] + Z.loc['B','orders']) / (Z.loc['A','visits'] + Z.loc['B','visits'])

z_value = (p_A - p_B) / (p_AB * (1 - p_AB) * (1/Z.loc['A','visits'] + 1/Z.loc['B','visits'])) ** (1/2)

distr = stats.norm(0, 1)  # стандартное нормальное распределение (среднее 0, стандартное отклонение 1)

alpha = 0.05
p_value = (1 - distr.cdf(abs(z_value))) * 2


# вывод результата

space()
print('p-value: {0:.4f}'.format(p_value))

if (p_value < alpha):
    print('отвергаем нулевую гипотезу: разница статистически значима при alpha={}'.format(alpha))
else:
    print('не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя при alpha={}'.format(alpha))
```

>Тест на "сырых" данных выявил статистически значимое различие в конверсии между группами. При этом конверсия за исследуемый период в **группе В** оказалась выше на 16%.  
Необходимо повторить тест на очищенных от аномалий данных.

### Результат теста

Для принятия решения о результате теста необходима дополнительная информация.  
Поэтому **построим дополнительные графики, используя очищенные от аномалий данные.**  
Код для графиков идентичен, вместо входной таблицы `orders` используется `orders_clean_99`.

Кроме того, построим суточный график относительной конверсии групп.

```python
# объединениение агрегированных orders_clean_99 и visits
aggregation = pd.merge(
                       (
                        orders_clean_99
                        .groupby(['group', 'date'])
                        .agg({'id_order':'nunique', 'id_user':'nunique', 'revenue':'sum'})
                        .rename(columns={'id_order':'orders', 'id_user':'buyers'})
                        .reset_index()
                       ),
                       visits,
                       on=['group', 'date'],
                       how='left'                          # для подстраховки
                      )

# создание кумулятивных данных (накопление по дате раздельно для групп)
aggregation_cumulative = (
                          aggregation
                          .set_index(['date','group'])
                          .sort_values(['date','group'])   # для подстраховки
                          .groupby(level=[1])              # группировка по полю ['group']
                          .cumsum()
                          .reset_index()
                         )
```

#### Относительное изменение кумулятивной конверсии группы B к группе A (очищенные данные)

```python
# расчет относительного изменения кумулятивной конверсии группы B к группе A

AB = pd.merge(
              aggregation_cumulative.query('group == "A"')[['date', 'conversion']],
              aggregation_cumulative.query('group == "B"')[['date', 'conversion']],
              on='date',
              how='outer',
              suffixes=['_A', '_B']
             )

AB['conversion_B_relative_A'] = AB['conversion_B'] / AB['conversion_A'] - 1

# график относительного изменения кумулятивной конверсии группы B к группе A

fig, ax = plt.subplots(figsize=(16,6), dpi=150)

ax.plot(AB['date'], AB['conversion_B_relative_A'], label='группа B', linewidth=5, alpha=0.4, color='red')
ax.axhline(y=0, linestyle='-', xmin=0.045, xmax=0.955, linewidth=5, alpha=0.4, color='blue', label='группа A')
ax.set_title('Относительное изменение кумулятивной конверсии группы B к группе A\n(очищенные данные)', fontsize=20, pad=25)
ax.set_xlabel('дата наблюдения', fontsize=12, labelpad=20)
ax.set_ylabel('дельта относительного изменения', fontsize=12, labelpad=20)
ax.legend(loc='lower right', fontsize=16)
ax.set_ylim(-0.2, 0.3)
ax.grid(True, linestyle='--')
ax.tick_params(labelsize=12)
ax.set_facecolor('#f9f9f9')

space()
plt.show()
```

>При ближайшем рассмотрении видно, что разрыв между кумулятивной конверсией групп постепенно сокращается. Стоит лишний раз напомнить, что линейное снижение кумулятивной величины говорит об ускоряющемся снижении величины в моменте.   
Сделать вывод о наличии тенденции постепенного выравнивания конверсии в группах не дает поведение графика в последние 2 дня: кумулятивная конверсия группы В резко пошла вверх. На этом данные заканчиваются.  

#### Суточный график относительной конверсии групп (очищенные данные)

```python
# расчет суточной относительной конверсии групп 
# если конверсия A > B, то y = -(A/B - 1)
# если конверсия B > A, то y = B/A - 1

aggregation['conversion'] = aggregation['buyers'] / aggregation['visitors']

AB = pd.merge(
              aggregation.query('group == "A"')[['date', 'conversion']],
              aggregation.query('group == "B"')[['date', 'conversion']],
              on='date',
              how='outer',
              suffixes=['_A', '_B']
             )

# данные для (y > 0)
AB['B_relative_A'] = AB['conversion_B'] / AB['conversion_A'] - 1
# данные для (y < 0)
AB['B_relative_A'] = AB['B_relative_A'].mask(
                                             AB['B_relative_A'] < 0,
                                             1 - AB['conversion_A'] / AB['conversion_B']
                                            )
                                
# график относительного изменения конверсии группы B к группе A

fig, ax = plt.subplots(figsize=(16,6), dpi=150)

ax.plot(AB['date'], AB['B_relative_A'], linewidth=5, alpha=0.2, color='black')

ax.bar(AB['date'], AB['B_relative_A'], color=['blue' if v < 0 else 'red' for v in AB['B_relative_A']], width=0.15, linewidth=5, alpha=0.4)

ax.axhline(y=0, linestyle='-', xmin=0.045, xmax=0.955, linewidth=2, color='black')
ax.set_title('Суточный график относительной конверсии групп\n(очищенные данные)', fontsize=20, pad=25)
ax.set_xlabel('дата наблюдения', fontsize=12, labelpad=20)
ax.set_ylabel('< лидирует A <     0     > лидирует B >    ', fontsize=12, labelpad=20)
ax.set_ylim(-2, 2.3)
ax.grid(True, linestyle='--')
ax.tick_params(labelsize=12)
ax.set_facecolor('#f9f9f9')

space()
plt.show()
```

>График можно разделить на 4 условные зоны:
>1. Короткий примерно равный начальный период.
>1. Период наращивания преимущества кумулятивной конверсии в группе В.
>1. Период примерно равной среднедневной конверсии, сопровождающийся резкими суточными колебаниями.
>1. Возврат к наращиванию  преимущества кумулятивной конверсии в группе В (возможно, этот период еще не окончился).
>
>В этом графике примечательно:
>1. Деление на явно выраженные периоды, особенно 2-й и 3-й. Возможно, это случайное совпадение.
>1. Лишь в 13 датах из 31 разница конверсий не достигла 50%. В остальные дни она была не только выше, но и часто превышала 100%.
>1. Группа А никогда не имела преимущества более 1 дня подряд (маленький хвостик 19 августа не считается).

#### Решение по результатам теста

Окончательный вывод о результатах теста базируется на очищенных от аномалий данных.

**Средний чек**

Между статистическим тестом и графиками сложился консенсус: разницы между группами нет. Решающее значение будет иметь показатель конверсии.

**Конверсия**

Статистика дает сигнал о более высокой конверсиии в **группе В**. Графики в целом подтверждают это, но в то же время на графиках видно постепенное сближение конверсий в группах. Эта картина еще более усложняется тем, что после продолжительного сближения в последние 2 дня разрыв вновь резко увеличился. В качестве решающего аргумента можно отметить, что группа А никогда не имела преимущества в конверсии более 1 дня подряд.

**Результат**

На данный момент **группа В** предпочтительнее благодаря более высокой конверсии. С одной стороны, продолжение теста может сопровождаться потерей части выручки. С другой, не факт, что более высокая конверсия в **группе В** сохранится надолго.
Учитывая, что даже если конверсии в группах сравняются, нет оснований предполагать, что конверсия в **группе В** установится ниже конверсии **группы А**.  
**<font color='blue'>Поэтому можно остановить тест и признать победу **группы В**.</font>

В качестве альтернативного решения можно продлить тест еще на 1-2 недели, чтобы попробовать выявить тенденцию относительной конверсии в группах.
